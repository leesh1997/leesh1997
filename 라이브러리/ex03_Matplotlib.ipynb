### 선 그래프

```python
# plot 함수 - 일직선으로 되어있는 Lineplot 그리기

y = [2,4,6]

plt.plot(y)   # 그래프 그리기
plt.show()    # 그래프 출력
```

```python
# 한 영역 안에 여러 그래프 그리기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, color='r', label='x')
plt.plot(y, color='k', label='y')
plt.plot(z, color='m', label='z')
plt.legend(loc='lower right')    # 범례 띄우기
plt.show()

# best
#    upper right
#    upper left
#    lower left
#    lower right
#    right
#    center left
#    center right
#    lower center
#    upper center
#    center
```

```python
plt.plot(x, color='r', label='엑스')
plt.plot(y, color='k', label='와이')
plt.plot(z, color='m', label='제트')
plt.xlabel('x축')
plt.ylabel('y축', rotation=0)
plt.title('x,y,z 그래프')
plt.legend()    # 범례 띄우기
plt.show()
```

#### 차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도를 알아보고 시각화

```python
data.head(2)
```

```python
# 1. 차대차 사건 중
# 2. (1)죽거나 다친 사람(사상자)이 많은 (2) 발생지 시도
# 3. 시각화
```

```python
acc=data[data['사고유형_대분류']=='차대차']
acc
```

```python
data['사고유형_대분류'][data['사고유형_대분류']=='차대차'].value_counts() # 불리언인덱싱이용
```

```python
acc['발생지시도'].unique()
```

```python
acc['사상자수'].value_counts()
```

```python
# 그룹화 = groupby( '그룹화 시킬 컬럼' ).sum()   mean도 가능(연산이나 함수 필요)
acc_g=acc.loc[:,['사상자수','발생지시도']].groupby('발생지시도').sum()
acc_g
```

```python
x=acc_g.index
y=acc_g['사상자수']
```

```python
# 시각화
plt.figure(figsize=(10,5)) #(가로,세로)
plt.bar(x,y, color = 'gray')
plt.title('차vs차 교통사고의 사상자 수',fontsize=25,color='lightgray')
plt.xlabel('지역', fontsize=13, color='lightblue')
plt.ylabel('사상자 수', fontsize=13, color='lightblue')
plt.show()
```

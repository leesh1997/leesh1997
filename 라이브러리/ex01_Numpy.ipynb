## 영화 평점 데이터 분석하기

```python
# 데이터 타입 변경
# dtype
data = np.loadtxt('ratings.txt', delimiter="::",dtype = 'int')
data
```

```python
# 전체 평점 평균 구하기
np.mean(data[:,2])
```

```python
# 각 사용자별 평점 평균 구하기
# 리스트에 [사용자, 사용자의 평점평균] 저장하기
# 1번 사용자의 평점 평균 구하기
num1=data[(data[:,0]==1)] # 전체 정보에서 id값이 1인 값만 출력
np.mean(num1[:,2])
num2=[]
for i in range(1,6041):
    num = data[(data[:,0]==i)]
    num2.append([i, np.mean(num[:,2])])
num2
```

```python
# 각 사용자별 평균 평점이 4점 이상인 사용자 구하기
user_list4 = []
for i in range(6040):
    if num2[i][1] >= 4:
        user_list4.append(num2[i])

user_list4
```
